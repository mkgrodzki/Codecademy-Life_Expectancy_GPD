# Introduction

This project will investigate the relationships between GPD of a country and it's citizens life expectancy.

The goals are to search through the data, clean it and then make good plots summarazing interesting insights.

This project will try to answer following questions:
+ Do life expactancy for each country only increase over the years?
+ Do GPD for each country only increase over the years?
+ Is GPD related with life expectancy?

```python
# Importing necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Loading the data and printing first 5 rows
df = pd.read_csv('all_data.csv')
df.head()
```

```python
# Cleaning the data
df = df.rename({"Life expectancy at birth (years)":"LEABY"}, axis = "columns")
```

```python
# Making general summary of the data
df.describe(include='all')
```

```python
# Printing countries included in the data
countries = df.Country.unique()
print(countries)
```

```python
# Analyzing all countries together
plt.figure(figsize=(9, 6))
for country in countries:
    sns.lineplot(data=df[df.Country == country], x = 'Year', y = 'GDP')
plt.legend(countries)
plt.title('Change of GPD over the years')
plt.show()
plt.clf()

plt.figure(figsize=(9, 6))
for country in countries:
    sns.lineplot(data=df[df.Country == country], x = 'Year', y = 'LEABY')
plt.legend(countries)
plt.title('Change of life expectancy over the years')
plt.show()
plt.clf()
```

```python
# Analyzing each country seperately
for country in countries:
    print (f"Analyzing {country}:")
    print(df[df.Country == country][['LEABY', 'GDP']].describe(include='all'))
    print()
```

```python
for country in countries:
    sns.lineplot(x = 'Year', y = 'GDP' , data=df[df.Country == country], color = 'r', legend=False)
    ax2 = plt.twinx()
    sns.lineplot(x = 'Year', y = 'LEABY' , data=df[df.Country == country], legend=False)
    plt.title(f'{country}')
    plt.show()
    plt.clf()
```

+ Do life expactancy for each country only increase over the years?

Life expactancy for each country increase over the years with a small exception for Zimbabwe in the early years of the 21st century.

+ Do GPD for each country only increase over the years?

Only China has increased it's GPD each year. Other countries have experienced significant growth over the 2000-2015 period, however, there are years in which GPD declined relatively to previous year.
+ Is GPD related with life expectancy?

There is a strong relationship between GPD and life expectancy of the country. However, some countries experience only growth in life expectancy, even though for some periods a decrease in GPD values can be observed (USA, Germany).
